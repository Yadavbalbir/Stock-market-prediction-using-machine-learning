```python
#import packages
import pandas as pd
import numpy as np

#to plot within notebook
import matplotlib.pyplot as plt
%matplotlib inline

#setting figure size
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20,10

#for normalizing data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

#read the file
df = pd.read_csv('AAPL.csv')

#print the head
print(df.describe())
```

```python
#reindexing
df.set_index(pd.DatetimeIndex(df['Date']), inplace=True)

#we are taking only the close price values
df = df[['Close']]

print(df)
```

```python
#plot
plt.figure(figsize=(16,8))
plt.plot(df['Close'], label='Close Price history')
```

Linear regression requires a series of assumptions to be made to be effective. One of these assumptions is that variables in the data are independent. This dictates that the residuals (difference between the predicted value and observed value) for any single variable aren’t related. For Time Series data this is often a problem since our observed values are longitudinal in nature—meaning they are observed values for the same thing, recorded in sequence. This produces a characteristic called autocorrelation which describes how a variable is somehow related to itself. The takeaway here is that our date values aren’t suitable as our independent variable and we need to come up with something else and use the close value as the independent variable. In order to cope with this problem we add technical indicators, in our case we are using Moving Averages.

```python
import pandas_ta

df.ta.ema(close='close', length=10, append=True)
print(df)
```

```python
#inspecting the first 10 rows
print(df.head(10))
```

```python
#the first few values for EMA_10 are NaN in order to tackle with this issue we are going to drop these rows,
#we could also have made them 0 but that's not a good option while predicting.
df=df.iloc[10:]
#inspecting the new datset
print(df.head(10))
```

```python
#inspecting how the EMA value tracks with the close prices
plt.figure(figsize=(16,8))
plt.plot(df['Close'], label='Close Price history')
plt.plot(df['EMA_10'], label='Close Price history')
```

```python
#!pip install sklearn
```

```python
from sklearn.model_selection import train_test_split
# Split data into testing and training sets
X_train, X_test, y_train, y_test = train_test_split(df[['Close']], df[['EMA_10']], test_size=0.2)

```

```python
#Test data
print(X_test.describe())
```

```python
#Training data
print(X_train.describe())
```

Now we need to train the model using the linear regression.


```python
from sklearn.linear_model import LinearRegression
#Regression Model
model = LinearRegression()
#Training the model
model.fit(X_train, y_train)
#Making predictions
y_pred = model.predict(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Now we have trained the model on 999 training samples and predicted the values
#We will now use the predicted values to calculate the mean squared error, the coefficient of determination and other errors
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print("Model Coefficients:", model.coef_)
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Coefficient of Determination:", r2_score(y_test, y_pred))
```

The value of the coefficient of determination suggests that the model fits our data well.

```python
#plotting the predicted vs real values graph
plt.scatter(X_test,y_test,color='magenta')
plt.plot(X_test,y_pred,color='black')
plt.legend(["Predicted Values", "Real Values"], loc ="upper left")
plt.show()
```
